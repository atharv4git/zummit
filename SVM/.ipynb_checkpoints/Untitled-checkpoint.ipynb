```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter('ignore')
```

```python
data = pd.read_csv('applicant.csv', names=[
        "Age","Workclass","fnlwgt","Education","Education-Num","Marital Status",
        "Occupation","Relationship","Race","Gender","Capital Gain","Capital Loss",
        "Hours per week","Country","Target"])
data.reset_index(inplace=True,drop=True)
data.head()
```

```python
g.map(sns.distplot,"Education-Num")
```

```python
num_cols = [i for i in data.columns if data[i].dtype!='O']
cat_cols = [i for i in data.columns if data[i].dtype=='O']
print(num_cols)
print(cat_cols)
```

```python
num_data = data[num_cols]
cat_data = data[cat_cols]
```

```python
fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(15, 12))
plt.subplots_adjust(hspace=0.5)
fig.suptitle("applicant - Distributions", fontsize=18, y=0.95)

for col, ax in zip(num_data.columns.to_list(), axs.ravel()):
    sns.distplot(num_data[col],ax=ax)

    ax.set_title(col.upper())
    ax.set_xlabel("")

plt.show()
```

```python
cat_cols
```

```python
fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(15, 12))
plt.subplots_adjust(hspace=1)
fig.suptitle("applicant - Countplots", fontsize=18, y=0.95)
for col, ax in zip(cat_data.columns.to_list(), axs.ravel()):
    sns.countplot(cat_data[col],ax=ax)
    ax.tick_params(axis='x', rotation=45)
    ax.set_title(col.upper())
    ax.set_xlabel("")

plt.show()
```

```python
import sklearn.preprocessing as preprocessing
le=preprocessing.LabelEncoder()
data['Occupation']=le.fit_transform(data['Occupation'].astype(str))
data.head()
```

```python
data['Target']=le.fit_transform(data['Target'].astype(str))
data.tail()
```

```python
from sklearn.model_selection import train_test_split
x=data[['Education-Num','Occupation']]
y=data['Target']
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.svm import SVC
classifier=SVC()
classifier.fit(xtrain,ytrain)
score=classifier.score(xtest,ytest)
print(score)
```

```python
data['Race']=le.fit_transform(data['Race'].astype(str))
data['Gender']=le.fit_transform(data['Gender'].astype(str))
data['Marital Status']=le.fit_transform(data['Marital Status'].astype(str))
data['Education']=le.fit_transform(data['Education'].astype(str))
```

```python
x=data[['Education-Num','Occupation','Age','Gender']]
y=data['Target']
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
classifier=SVC()
classifier.fit(xtrain,ytrain)
score=classifier.score(xtest,ytest)
print(score)
```

```python
classifier =SVC(kernel='rbf', C=1.0)
classifier.fit(xtrain,ytrain)
score=classifier.score(xtest,ytest)
print(score)
```

```python
classifier =SVC(kernel='linear', C=10.0)
classifier.fit(xtrain,ytrain)
score=classifier.score(xtest,ytest)
print(score)
```
